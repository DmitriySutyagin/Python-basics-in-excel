#### **Задача 3**  
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.  
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(10,8))
sns.set(font_scale = 1.0)
sns.heatmap(corr_matrix, annot= True, linewidths= 5, cmap= 'coolwarm')
plt.title('Correlation matrix');
```

Вывод: На данной тепловой карте видно, что нету ни одной сильной обратной зависомости. Есть только средние и слабые обратные зависомости.А также присутвуют в большом количестве сильные и средние прямые зависимости признаком. Далее разберем некоторые из них подробнее.

```python
sns.jointplot(x=df['sqft_living'], y=df['sqft_above'], kind='reg');
```
